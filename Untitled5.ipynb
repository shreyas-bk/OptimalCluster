```python
from OptimalCluster.opticlust import Optimal
opt = Optimal({'max_iter':350})
```

```python
import pandas as pd
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 6, 6
x, y = make_blobs(1000, n_features=2, centers=3)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
df = pd.DataFrame(x,columns=['A','B'])
```

```python
opt.elbow(df)
```

```python
opt.elbow(df,display=True,visualize=True)
```

```python
x, y = make_blobs(1000, n_features=2, centers=6)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
df = pd.DataFrame(x,columns=['A','B'])
```

```python
opt.elbow(df,display=True,visualize=True,method='lin')
```

```python
x, y = make_blobs(1000, n_features=3, centers=8)
plt.scatter(x[:, 0], x[:, 1])
plt.show()
df = pd.DataFrame(x,columns=['A','B','C'])
```
